##**Allen NLP Model**

```python
import pandas as pd
df = pd.read_csv('/content/NLP_Q_T5.csv')
```

```python
df
```

```python
from allennlp.predictors.predictor import Predictor
import allennlp_models.rc
Allen_Answers = []
predictor = Predictor.from_path("https://storage.googleapis.com/allennlp-public-models/bidaf-elmo.2021-02-11.tar.gz")
for item in range(df['Context'].size):
  Allen_Answers.append(predictor.predict(
      passage=df['Context'][item],
      question=df['Question Using T5 Model'][item])['best_span_str'])
```

```python
Allen_Answers
```

```python
df['Allen_Answers'] = Allen_Answers
```

```python
df
```

```python
df.to_csv('NLP_Q_T5_ALLEN.csv',index=False)
```
