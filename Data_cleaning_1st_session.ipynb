Remove Arabic Words

```python
def remove_arabic(arabic):
    import re
    arabic = re.sub("[^a-zA-Z\,\.\+]+"," ",arabic)
    return arabic
```

```python
contents = remove_arabic(contents)
```

Remove Specific Puncituations 

```python
punctuations = "!$%&()*-/:;<=>?@[\]^_{|}~"
```

```python
def remove_punctuation(text):
    punctuationfree="".join([i for i in text if i not in punctuations])
    return punctuationfree
```

```python
contents = remove_punctuation(contents)
```
